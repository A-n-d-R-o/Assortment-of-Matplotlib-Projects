```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# projectile motion visualisation
def projectile(u: float, theta: float):

    theta = theta * np.pi / 180
    
    h_range = u**2 * np.sin(2*theta) / 9.8
    time = abs(h_range / (u * np.cos(theta)))
    max_height = abs(0.5 * u * np.sin(theta) * time)

    x = np.linspace(0, h_range, 1000)
    y = [x_i * np.tan(theta) - (9.8 * x_i**2) / (2 * (u * np.cos(theta))**2) for x_i in x]
        
    print(f'Max height = {max_height:.3g} m')
    print(f'Range (displacement) = {h_range:.3g} m')
    print(f'Time of flight = {time:.3g} s')
            
    plt.plot(x, y)
    plt.axis('equal')
    plt.axhline(0, color='black')
    plt.axvline(0, color='black', linestyle=':')
    plt.xlabel('Distance $m$')
    plt.ylabel('Height $m$')
```

```python
projectile(100, 45)
```
