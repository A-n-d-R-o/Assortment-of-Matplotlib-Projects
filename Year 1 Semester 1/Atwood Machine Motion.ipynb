```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Atwood machine motion graphs
def Atwood(m1, m2):
    
    ratio = (m1 - m2) / (m1 + m2)
    
    time = np.linspace(0, 3, 1000)
    acceleration = ratio * -9.8
    velocity = acceleration * time
    position = 0.5 * velocity * time**2
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
    
    ax[0].plot(time, position, color='red', label='position')
    ax[0].plot(time, velocity, color='green', label='velocity')
    ax[0].axhline(acceleration, color='blue', label='acceleration')
    ax[0].set_xlim(0, 3)
    ax[0].set_title(r'$m_1$ motion')
    ax[0].legend()
    
    ax[1].plot(time, -position, color='red', label='position')
    ax[1].plot(time, -velocity, color='green', label='velocity')
    ax[1].axhline(-acceleration, color='blue', label='acceleration')
    ax[1].set_xlim(0, 3)
    ax[1].set_title(r'$m_2$ motion')
    ax[1].legend();
```

```python
Atwood(m1=2, m2=3)
```
