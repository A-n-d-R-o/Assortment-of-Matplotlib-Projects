```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# vector addition and subtraction
def vector(v_1: list, sign: str, v_2: list):

    v_1 = np.array([v_1[0], v_1[1]])
    v_2 = np.array([v_2[0], v_2[1]])
    
    if sign == '+':
        v_3 = np.add(v_1, v_2)
    elif sign == '-':
        v_3 = np.subtract(v_1, v_2)

    print(f'{v_1} {sign} {v_2} = {v_3}')
        
    vectors = [v_1, v_2, v_3]

    fig, ax = plt.subplots()
    for v, colour in zip(vectors, ['r', 'r', 'b']):
        ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=colour)
    
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')

    all_x = [v[0] for v in vectors] + [0]
    all_y = [v[1] for v in vectors] + [0]
    x_min, x_max = min(all_x), max(all_x)
    y_min, y_max = min(all_y), max(all_y)
    
    padding = 1
    ax.set_xlim(x_min - padding, x_max + padding)
    ax.set_ylim(y_min - padding, y_max + padding)
    ax.set_aspect('equal')
    ax.set_axisbelow(True)
    
    plt.grid()
    plt.show()
```

```python
vector([4, -1], '+', [2, 5])
```
