```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# damping diagram
def damp(time, x_m, b, m, omega, phi):

    t = np.linspace(0, time, 1000)
    envelope = x_m * np.exp(-b * t / (2 * m))
    oscillation = np.cos(omega * t + phi)
    curve = envelope * oscillation

    max_val = 1.1*max(envelope)

    plt.figure(figsize=[10, 5])
    plt.grid(True)
    
    plt.ylim([-max_val, max_val])
    
    plt.plot(t, curve, color='blue', label='Damped oscillation')
    plt.plot(t, envelope, linestyle='--', color='red', label='Envelope curve')
    plt.plot(t, -envelope, linestyle='--', color='red')

    plt.title('Damped Harmonic Motion')
    plt.ylabel('Displacement from equilibrium')
    plt.xlabel('Time')
    plt.legend();
```

```python
damp(time=100, x_m=5, b=0.5, m=4, omega=2, phi=0)
```
