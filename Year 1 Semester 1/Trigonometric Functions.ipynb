```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
# trig graphs
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
x_arc = np.linspace(-1, 1, 1000)
x_arc2 = np.concatenate([np.linspace(-2*np.pi, -1, 500)[:-1], np.linspace(1, 2*np.pi, 500)[1:]])
x_hyp = np.linspace(-np.pi, np.pi, 1000)
x_arc3 = np.linspace(1, 3*np.pi, 1000)
x_arc4 = np.linspace(0, 1, 1000)

sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
tanx[:-1][abs(np.diff(tanx)) > 10] = np.nan

asinx = np.asin(x_arc)
acosx = np.acos(x_arc)
atanx = np.atan(x)

cscx = 1/np.sin(x)
cscx[:-1][abs(np.diff(cscx)) > 10] = np.nan
secx = 1/np.cos(x)
secx[:-1][abs(np.diff(secx)) > 10] = np.nan
cotx = 1/np.tan(x)
cotx[:-1][abs(np.diff(cotx)) > 10] = np.nan

arccscx = np.asin(1/x_arc2)
arccscx[:-1][abs(np.diff(arccscx)) > 1] = np.nan
arcsecx = np.acos(1/x_arc2)
arcsecx[:-1][abs(np.diff(arcsecx)) > 1] = np.nan
arccotx = 0.5*np.pi - np.atan(x)

sinhx = np.sinh(x_hyp)
coshx = np.cosh(x_hyp)
tanhx = np.tanh(x_hyp)

arcsinhx = np.arcsinh(x)
arccoshx = np.arccosh(x_arc3)
arctanhx = np.arctanh(x_arc[1:-1])

cschx = 1/np.sinh(x_hyp)
cschx[:-1][abs(np.diff(cschx)) > 10] = np.nan
sechx = 1/np.cosh(x)
cothx = 1/np.tanh(x_hyp)
cothx[:-1][abs(np.diff(cothx)) > 10] = np.nan

def arccsch(x):
    return np.log(1/x+np.sqrt(1+x**2)/np.abs(x))
def arcsech(x):
    return np.log(1/x+np.sqrt(1/x**2-1))
def arccoth(x):
    return 0.5*np.log((x+1)/(x-1))
    
arccschx = arccsch(x)
arccschx[:-1][abs(np.diff(arccschx)) > 10] = np.nan
arcsechx = arcsech(x_arc4[1:])
arccothx = arccoth(x_arc2)
arccothx[:-1][abs(np.diff(arccothx)) > 2] = np.nan
```

```python
fig, ax = plt.subplots(nrows=8, ncols=3, figsize=(10,14))

# regular trig functions

ax[0,0].plot(x, sinx)
ax[0,0].set_title(r'$y = \sin\theta$')

ax[0,1].plot(x, cosx)
ax[0,1].set_title(r'$y = \cos\theta$')

ax[0,2].plot(x, tanx)
ax[0,2].set_title(r'$y = \tan\theta$')
ax[0,2].set_xlim(-1.5*np.pi, 1.5*np.pi)
ax[0,2].set_ylim(-10, 10)

# inverse regular trig funnctions

ax[1,0].plot(x_arc, asinx)
ax[1,0].set_title(r'$\theta = \sin^{-1} y$')

ax[1,1].plot(x_arc, acosx)
ax[1,1].set_title(r'$\theta = \cos^{-1} y$')

ax[1,2].plot(x, atanx)
ax[1,2].set_title(r'$\theta = \tan^{-1} y$')

# reciprocal trig functions

ax[2,0].plot(x, cscx)
ax[2,0].set_title(r'$y = \csc\theta$')
ax[2,0].set_xlim(-1.5*np.pi, 1.5*np.pi)
ax[2,0].set_ylim(-10, 10)

ax[2,1].plot(x, secx)
ax[2,1].set_title(r'$y = \sec\theta$')
ax[2,1].set_xlim(-1.5*np.pi, 1.5*np.pi)
ax[2,1].set_ylim(-10, 10)

ax[2,2].plot(x, cotx)
ax[2,2].set_title(r'$y = \cot\theta$')
ax[2,2].set_xlim(-1.5*np.pi, 1.5*np.pi)
ax[2,2].set_ylim(-10, 10)

# inverse reciprocal trig functions

ax[3,0].plot(x_arc2, arccscx)
ax[3,0].set_title(r'$\theta = \csc^{-1} y$')

ax[3,1].plot(x_arc2, arcsecx)
ax[3,1].set_title(r'$\theta = \sec^{-1} y$')

ax[3,2].plot(x, arccotx)
ax[3,2].set_title(r'$\theta = \cot^{-1} y$')

# hyperbolic trig functions

ax[4,0].plot(x_hyp, sinhx)
ax[4,0].set_title(r'$y = \sinh x$')

ax[4,1].plot(x_hyp, coshx)
ax[4,1].set_title(r'$y = \cosh x$')

ax[4,2].plot(x_hyp, tanhx)
ax[4,2].set_title(r'$y = \tanh x$')

# inverse hyperbolic trig functions

ax[5,0].plot(x, arcsinhx)
ax[5,0].set_title(r'$x = \sinh^{-1} y$')

ax[5,1].plot(x_arc3, arccoshx)
ax[5,1].set_title(r'$x = \cosh^{-1} y$')

ax[5,2].plot(x_arc[1:-1], arctanhx)
ax[5,2].set_title(r'$x = \tanh^{-1} y$')

# reciprocal hyperbolic trig functions

ax[6,0].plot(x_hyp, cschx)
ax[6,0].set_title(r'$y = \mathrm{csch}\, x$')

ax[6,1].plot(x, sechx)
ax[6,1].set_title(r'$y = \mathrm{sech}\, x$')

ax[6,2].plot(x_hyp, cothx)
ax[6,2].set_title(r'$y = \mathrm{coth}\, x$')

# inverse reciprocal hyperbolic trig functions

ax[7,0].plot(x, arccschx)
ax[7,0].set_title(r'$x = \mathrm{csch}^{-1}\, y$')

ax[7,1].plot(x_arc4[1:], arcsechx)
ax[7,1].set_title(r'$x = \mathrm{sech}^{-1}\, y$')

ax[7,2].plot(x_arc2, arccothx)
ax[7,2].set_title(r'$x = \mathrm{coth}^{-1}\, y$')

plt.tight_layout();
plt.savefig('trigs.png')
```
