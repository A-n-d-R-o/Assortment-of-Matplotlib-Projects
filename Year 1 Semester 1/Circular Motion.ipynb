```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# uniform circular motion
def circular_motion(position, max_position=1000):
    
    theta = np.linspace(0, 2*np.pi, max_position)
    all_x = [np.cos(theta_i) for theta_i in theta]
    all_y = [np.sin(theta_i) for theta_i in theta]
    
    x_pos = all_x[position]
    y_pos = all_y[position]
    
    x_vel = -np.sin(theta[position])
    y_vel = np.cos(theta[position])
    
    x_acc = -np.cos(theta[position])
    y_acc = -np.sin(theta[position])
    
    fig, ax = plt.subplots()
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    ax.set_aspect('equal')
    
    plt.plot(all_x, all_y, zorder=1, label='path of motion')
    plt.scatter(x_pos, y_pos, color='red', zorder=3, label='position')
    ax.quiver(x_pos, y_pos, x_vel, y_vel, scale=5, color='orange', zorder=2, label='velocity')
    ax.quiver(x_pos, y_pos, x_acc, y_acc, scale=5, color='green', zorder=2, label='acceleration')
    plt.legend();
```

```python
circular_motion(125)
```
